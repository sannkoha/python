```python
# PROBLEM NUMBER 1
# returns the mean of list
#   x
def mean(x):

    # result holds the result
    # set result to 0
    # count holds the number
    #   of valid elements
    #   of x
    # set count to 0
    # iterate each element
    #   of x using element
    #   as element
    result = 0
    count = 0
    for element in x:

        # if element is int or float
        # add result to element
        #   and update result
        # increment count by 1
        if isinstance(element, int) or \
            isinstance(element, float):
            result = result + element
            count = count + 1

    # if count is 0
    #   return none
    # otherwise count is
    #   greater than 0
    #   return result divided
    #     by count
    if count == 0:
        return None
    else:
        return result / count


# PROBLEM NUMBER 2
# returns the index of the
#   maximum element of list
#   x
def imax(x):

    # result holds the result
    # set result to 0
    # count holds the number
    #   of valid elements
    #   of x
    # set count to 0
    # maximum holds the maximum
    #   value of x
    # iterate each element
    #   of x using index
    #   as index
    result = 0
    count = 0
    maximum = 0
    for index in range(len(x)):

        # if element is int or float
        # if count is 0
        #   we have our first
        #     int or float
        #   set maximum to x of index
        #   set result to index
        # otherwise this is not
        #   our first int or float
        #   if x of index is greater
        #     than maximum
        #     set maximum to x of index
        #     set result to index
        # increment count by 1
        if isinstance(x[index], int) or \
            isinstance(x[index], float):
            if count == 0:
                maximum = x[index]
                result = index
            elif x[index] > maximum:
                maximum = x[index]
                result = index
            count = count + 1

    # if count is 0
    #   return none
    # otherwise count is
    #   greater than 0
    #   return result
    if count == 0:
        return None
    else:
        return result


# PROBLEM NUMBER 3
# returns the mode of list x
# it ALWAYS return a list
#   in order to accommodate
#   multiple modes
def mode(x):

    # dictionary holds the
    #   frequency table of x
    #   where element of x is
    #     the key
    #   and the frequency its
    #     value
    # iterate each element
    #   of x
    dictionary = {}
    for element in x:

        # if element is already
        #   in the dictionary
        #   increment it's value
        #     by 1
        # otherwise element
        #   does not exist in
        #   the dictionary
        #   add a key value pair
        #     with element as key
        #     and 1 as value
        if element in dictionary:
            dictionary[element] = dictionary[element] + 1
        else:
            dictionary[element] = 1

    # result holds the list
    #   of mode/modes
    # set result as an empty list
    # iterate each key and value
    #   pair of the dictionary
    result = []
    for key, value in dictionary.items():

        # if we have our first
        #   key value pair
        #   add key to result
        # else if value is greater than
        #   the element in result
        #   reset result
        #   add value as the first
        #     element of result
        # else if value is equal
        #   to the element in result
        #   we have multiple modes
        #   add key to result
        if len(result) == 0:
            result.append(key)
        elif value > result[0]:
            result = []
            result.append(key)
        elif value == result[0]:
            result.append(key)

    # return result
    return result


# PROBLEM NUMBER 4
# returns a list of tokenized
#   words separated by
#   whitespaces with each
#   word suitable for
#   caseless comparisons
#   (I converted them to lower
#   case) and comprised
#   only of alphanumeric characters
def tokenize(s):

    # result holds the result
    # set result as an empty
    #   list
    # call split of s
    #   and save list to
    #   words
    # iterate each word in words
    #   using word as the element
    result = []
    words = s.split()
    for word in words:

        # convert word to lowercase
        #   so that we can do caseless
        #   comparisons
        # string is a word with
        #   having only
        #   alphanumeric character
        # set string as an empty string
        # iterate each character of
        #   the word
        word = word.lower()
        string = ""
        for character in word:

            # if character is either a letter
            #   or a number
            #   add character to string
            if (character >= 'a' and character <= 'z') or \
                (character >= '0' and character <= '9'):
                string = string + character

        # if string is not empty
        #   add string to result
        if len(string) > 0:
            result.append(string)

    # return result
    return result


# PROBLEM NUMBER 5
# returns a dictionary
#   with alphanumeric only
#   keys containing
#   unique words of s
#   and values it's
#   corresponding frequencies
def count_words(s):

    # result holds the result
    # set result as an empty dictionary
    # call tokenize with s
    #   as parameter
    #   and save result to a_list
    # iterate each element of
    #   a_list using element
    #   as element
    result = {}
    a_list = tokenize(s)
    for element in a_list:


        # if element is already
        #   in result
        #   increment it's value
        #     by 1
        # otherwise element
        #   does not exist in
        #   result
        #   add a key value pair
        #     with element as key
        #     and 1 as value
        if element in result:
            result[element] = result[element] + 1
        else:
            result[element] = 1

    # return result
    return result


print(mean([1, 2, 3, 4, 5]))
print(imax([1, 2, 3, 4, 5]))
print(mode([1, 3, 3, 4, 5,5]))
#print(mode([9, 8, 7, 13,13]))
print(tokenize("The quick brown fox jumped over the lazy dog."))
print(count_words("The quick brown fox jumped over the lazy dog."))
```
