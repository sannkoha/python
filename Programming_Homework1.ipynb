```python
# Solution 1: Returns the mean in a list

def mean(x):

    result = 0 # result holds the result / # Let's initialize result to 0
    count = 0 # count holds the number of valid elements of x /  # Let's initialize to 0
    for element in x:  # let's iterate each element of x using element as element

       
        #if element is int or float add result to element and update result
         if isinstance(element, int) or \
           isinstance(element, float):
           result = result + element
           count = count + 1

    # if count is 0
    #   return none
    # otherwise count is
    #   greater than 0
    #   return result divided
    #     by count
    if count == 0:
        return None
    else:
        return result / count

#Solution 2 : returns the index of the maximum element of list x

def imax(x):

    # result holds the result
    # Let's initialize result to 0
    # count holds the number of valid elements of x
    # Let's initialize count to 0
    # maximum holds the maximum value of x
    # iterate each element of x using index as index
    result = 0
    count = 0
    maximum = 0
    for index in range(len(x)):

        # if element is int or float
        # if count is 0, we have our first int or float
        # initialize maximum to x of index
        # initialize result to index otherwise this is not our first int or float
        # if x of index is greater than maximum
        # initialize maximum to x of index
        # initialize result to index
        # increment count by 1
        if isinstance(x[index], int) or \
            isinstance(x[index], float):
            if count == 0:
                maximum = x[index]
                result = index
            elif x[index] > maximum:
                maximum = x[index]
                result = index
            count = count + 1

    # if count is 0
    #   return none
    # otherwise count is
    #   greater than 0
    #   return result
    if count == 0:
        return None
    else:
        return result
    
# Solution 3: returns the mode of list x

def mode(x):

    # dictionary holds the frequency table of x
    # where element of x is the key and the frequency its value
    # iterate each element of x
    
    dictionary = {}
    for element in x:

        
        
       
        
        # if element is already in the dictionary increment it's value by 1
        if element in dictionary:
            dictionary[element] = dictionary[element] + 1
        else: # otherwise element does not exist in the dictionary
            dictionary[element] = 1  # add a key value pair with element as key and 1 as value

    
   
    
    result = [] # result holds the list of mode/modes.  # set result as an empty list
    for key, value in dictionary.items():  # iterate each key and value pair of the dictionary

       
       
        
        if len(result) == 0:  # if we have our first key value pair add key to result
            result.append(key)
        elif value > result[0]:  # else if value is greater than the element in result reset result add value as the first element of result
            result = []
            result.append(key)
        elif value == result[0]: # else if value is equal to the element in result we have multiple modes add key to result
            result.append(key)

    # return result
    return result

# Solution 4: # returns a list of tokenized words 

# only of alphanumeric characters
def tokenize(s):

    
    
    
    result = [] # result holds the result set result as an empty list
    words = s.split() # call split of s and save list to words
    for word in words: # iterate each word in words using word as the element

        
        
         
        
        word = word.lower() # convert word to lowercase so that we can do caseless comparisons
        string = ""  # string is a word with only alphanumeric characters. set string as an empty string
        for character in word:  # iterate each character of the word

            # if character is either a letter or a number
            #  add character to string
            if (character >= 'a' and character <= 'z') or \
                (character >= '0' and character <= '9'):
                string = string + character

        # if string is not empty
        #   add string to result
        if len(string) > 0:
            result.append(string)

    # return result
    return result

# Solution 5: returns a dictionary with alphanumeric only keys containing unique words 

def count_words(s):

   
    
    
    
    result = {}   # result holds the result. # set result as an empty dictionary
    a_list = tokenize(s) # call tokenize with s as parameter and save result to a_list
    for element in a_list: # iterate each element of a_list using element as element


      
        
        if element in result:
            result[element] = result[element] + 1   # if element is already in result increment it's value by 1
        else:
    # otherwise element does not exist in result add a key value pair with element as key and 1 as value
            result[element] = 1

    # return result
    return result

# Let's test the functions
print(mean(["sss", "bbb", "aaa", "efg", "cdf", "abc"]))
print(imax([6, 9, 8, 7, 13]))
print(mode([6, 9, 8, 7, 13]))
print(tokenize("Twinkle twinkle litte start how I wonder what you are."))
print(count_words("Up above the world so high like a diamond in the sky."))
```
